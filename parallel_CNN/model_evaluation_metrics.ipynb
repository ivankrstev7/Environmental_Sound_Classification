```python
dataset = 'esc50' # 'esc10' or 'esc50'

# esc10

# model_dir = "./drive/MyDrive/HDA_project/models3_esc10/model.hdf5"
# model_dir = "./drive/MyDrive/HDA_project/models4_esc10/model.hdf5"

# esc50

# model_dir = "./drive/MyDrive/HDA_project/models4/model_5_1_noaug.hdf5"
model_dir = "./drive/MyDrive/HDA_project/models4/model_5_1_example.hdf5"
```

```python
model = load_model(model_dir)

test_data = np.load('drive/MyDrive/HDA_project/esc_melsp_test.npz')

if dataset == 'esc10':
  classes = 10
  esc10_ids = [ 0,  1, 10, 11, 12, 20, 21, 38, 40, 41]
  indexes = np.isin(test_data["y"], esc10_ids)

  x_test = test_data["x"][indexes]
  y_test = test_data["y"][indexes]

  # replace indexes in targets
  for i, index in enumerate(esc10_ids):
    y_test[y_test == index] = i

elif dataset == 'esc50':
  classes = 50
  x_test = test_data["x"]
  y_test = test_data["y"]
else:
  raise AssertionError(f'dataset "{dataset}" is not known. Please try "esc10" or "esc50".')

y_test = keras.utils.to_categorical(y_test, classes)
```

```python
x_test, x_val, y_test, y_val= model_selection.train_test_split(x_test, y_test, test_size=3/5, random_state=1)

print(len(x_test))
```

```python
evaluation = model.evaluate(x_test, y_test)
print(evaluation)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Assuming model.predict_classes(x_test) returns the predicted classes
y_pred = model.predict(x_test)

y_test_ = y_test.argmax(axis=1)
y_pred_ = y_pred.argmax(axis=1)

# Calculate accuracy, precision, recall, and F1 score
accuracy = accuracy_score(y_test_, y_pred_)
precision = precision_score(y_test_, y_pred_, average = 'weighted')
recall = recall_score(y_test_, y_pred_, average = 'weighted')
f1 = f1_score(y_test_, y_pred_, average = 'weighted')

print("Accuracy:", round(accuracy,3))
print("Precision:", round(precision,3))
print("Recall:", round(recall,3))
print("F1 Score:", round(f1,3))

print(round(accuracy,3), round(precision,3), round(recall,3), round(f1,3), sep = ', ')
```
